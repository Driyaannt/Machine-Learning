```python
import pandas as pd

# Baca dataset dari file CSV
data = pd.read_csv('wbc.csv') 

# Pilih kolom-kolom yang dapat digunakan
kolom_dapat_digunakan = ['diagnosis', 'radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean', 'smoothness_mean', 'compactness_mean', 'concavity_mean', 'concave points_mean', 'symmetry_mean', 'fractal_dimension_mean']

# Buat DataFrame baru hanya dengan kolom-kolom yang dapat digunakan
data_dapat_digunakan = data[kolom_dapat_digunakan]

# Tampilkan DataFrame baru
# print(data_dapat_digunakan)

df = data_dapat_digunakan
df.head(100)

```

```python

from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()

# Melakukan encoding pada kolom 'diagnosis'
data['diagnosis_encoded'] = label_encoder.fit_transform(data['diagnosis'])

# Tampilkan DataFrame dengan kolom 'diagnosis_encoded'
# print(data[['diagnosis', 'diagnosis_encoded']])

df = data[['diagnosis', 'diagnosis_encoded']]
df.head(100)
```

```python
from sklearn.preprocessing import StandardScaler

# Pilih hanya kolom-kolom yang memiliki nilai numerik
kolom_numerik = data.select_dtypes(include=['float64', 'int64'])

# Inisialisasi StandardScaler
scaler = StandardScaler()

# Lakukan standarisasi pada kolom-kolom numerik
data[kolom_numerik.columns] = scaler.fit_transform(kolom_numerik)

# Tampilkan DataFrame dengan kolom yang sudah di-standarisasi
df = data
df.head(100)
```

```python
from sklearn.model_selection import train_test_split

# Pilih kolom-kolom yang akan digunakan sebagai fitur (X) dan target (y)
X = data[['radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean', 'smoothness_mean', 'compactness_mean', 'concavity_mean', 'concave points_mean', 'symmetry_mean', 'fractal_dimension_mean']]
y = data['diagnosis']

# Lakukan stratified split data dengan rasio 80:20
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

# Tampilkan ukuran dari set data latih dan uji
print("Jumlah data latih:", len(X_train))
print("Jumlah data uji:", len(X_test))
```
